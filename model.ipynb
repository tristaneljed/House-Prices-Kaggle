```python
# Creating dummies
```

```python
train['enc_street'] = pd.get_dummies(train.Street, drop_first=True)
test['enc_street'] = pd.get_dummies(train.Street, drop_first=True)
```

```python
condition_pivot = train.pivot_table(index='SaleCondition',
                                    values='SalePrice', aggfunc=np.median)
condition_pivot.plot(kind='bar', color='blue')
plt.xlabel('Sale Condition')
plt.ylabel('Median Sale Price')
plt.xticks(rotation=0)
plt.show()
```

```python
# Notice that Partial has a significantly higher Median Sale Price than the others. 
# We will encode this as a new feature. We select all of the houses where SaleCondition is equal to Patrial and assign the value 1, 
# otherwise assign 0.
```

```python
def encode(x): return 1 if x == 'Partial' else 0
train['enc_condition'] = train.SaleCondition.apply(encode)
test['enc_condition'] = test.SaleCondition.apply(encode)
```

```python
condition_pivot = train.pivot_table(index='enc_condition', values='SalePrice', aggfunc=np.median)
condition_pivot.plot(kind='bar', color='blue')
plt.xlabel('Encoded Sale Condition')
plt.ylabel('Median Sale Price')
plt.xticks(rotation=0)
plt.show()
```

```python
# Dealing with missing values
```

```python
data = train.select_dtypes(include=[np.number]).interpolate().dropna() 
```

```python
# Check if the all of the columns have 0 null values.
sum(data.isnull().sum() != 0)
```

```python
# Step 3 : Build a linear model
```

```python
y = np.log(train.SalePrice)
X = data.drop(['SalePrice', 'Id'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
                                    X, y, random_state=42, test_size=.33)
```

```python
# Begin modelling
```

```python
from sklearn import linear_model
lr = linear_model.LinearRegression()
```

```python
model = lr.fit(X_train, y_train)
```

```python
# Evaluate the performance and visualize results
```

```python
print ("R^2 is: \n", model.score(X_test, y_test))
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
print ('RMSE is: \n', mean_squared_error(y_test, predictions))
```

```python
actual_values = y_test
plt.scatter(predictions, actual_values, alpha=.75,
            color='b') #alpha helps to show overlapping data
plt.xlabel('Predicted Price')
plt.ylabel('Actual Price')
plt.title('Linear Regression Model')
plt.show()
```

```python
# Try to improve the model
```

```python
# Weâ€™ll next try using Ridge Regularization to decrease the influence of less important features. 
# Ridge Regularization is a process which shrinks the regression coefficients of less important features.
```

```python
for i in range (-2, 3):
    alpha = 10**i
    rm = linear_model.Ridge(alpha=alpha)
    ridge_model = rm.fit(X_train, y_train)
    preds_ridge = ridge_model.predict(X_test)
    
    plt.scatter(preds_ridge, actual_values, alpha=.75, color='b')
    plt.xlabel('Predicted Price')
    plt.ylabel('Actual Price')
    plt.title('Ridge Regularization with alpha = {}'.format(alpha))
    overlay = 'R^2 is: {}\nRMSE is: {}'.format(
                    ridge_model.score(X_test, y_test),
                    mean_squared_error(y_test, preds_ridge))
    plt.annotate(s=overlay,xy=(12.1,10.6),size='x-large')
    plt.show()
```

```python
# These models perform almost identically to the first model. In our case, adjusting the alpha did not substantially 
# improve our model. As you add more features, regularization can be helpful. Repeat this step after youâ€™ve added more features.
```
